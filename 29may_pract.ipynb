```python
ndf = pd.read_csv('sample_data/california_housing_train.csv')
# Изобразите отношение households к population с помощью точечного графика
sns.scatterplot(data = ndf, x = 'households', y = 'population')
```

```python
# Визуализировать longitude по отношения к median_house_value, используя линейный график
sns.relplot(x='longitude', y='median_house_value', kind='line', data=ndf)
```

```python
# Представить гистограмму по housing_median_age
sns.histplot(data=ndf, x='housing_median_age')
```

```python
# Подключение таблицы с пингвинами
penguins = sns.load_dataset("penguins")
penguins.head()
```

```python
# Точечный график отношения bill_length_mm к bill_depth_mm
sns.scatterplot(data = penguins, x='bill_length_mm', y='bill_depth_mm', hue='sex', size=penguins['body_mass_g'])
```

```python
# Точечный график отношения body_mass_g к flipper_length_mm
sns.scatterplot(data = penguins, x='body_mass_g', y='flipper_length_mm', hue='sex', size=6)
```

```python
# Точечный график отношения bill_depth_mm к bill_length_mm
sns.scatterplot(data = penguins, x='bill_depth_mm', y='bill_length_mm', hue='body_mass_g', size=4)
```

```python
lst = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']
g = sns.PairGrid(penguins[lst])
g.map(sns.scatterplot)
```

```python
# Гистограмма bill_depth_mm
sns.histplot(data=penguins, x='bill_depth_mm')
```

```python
# Гистограмма bill_length_mm
sns.histplot(data=penguins, x='bill_length_mm')
```

```python
# Гистограмма flipper_length_mm
sns.histplot(data=penguins, x='flipper_length_mm')
```
